```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
edata = pd.read_excel('publichousesandbarsbylocalauthority20012019.xls',
                     sheet_name='Pubs size employment LA',header=1,skiprows=1,skipfooter=39)
etable = edata.drop([0])
```

```python
uk_f10 = etable.iloc[0,3:]
uk_m10 = etable.iloc[1,3:]
uk_any = etable.iloc[2,3:]
```

```python
years = uk_f10.index.values
```

```python
fig = plt.figure()  # Create a figure and an axes
ax1 = fig.add_subplot(2,1,1)
ax1.plot(years,uk_f10,label="Fewer than 10 emps")
ax1.plot(years,uk_m10,label="More than 10 emps")
ax1.plot(years,uk_any,label="Any number of emps")
ax1.set_xticks([2000,2005,2010,2015,2020])
ax1.set_xlabel('Year')
ax1.set_ylabel('People employed')
ax1.set_title('Total employment in Public Houses and Bars, UK')
ax1.legend()
```

```python
ndata = pd.read_excel('publichousesandbarsbylocalauthority20012019.xls',sheet_name='Pubs size LA',
                     header=1,skiprows=1,skipfooter=21)
ntab = ndata.drop([0])
```

```python
nuk_f10 = ntab.iloc[0,3:] # n refers to the number of people employed
nuk_m10 = ntab.iloc[1,3:]
nuk_any = ntab.iloc[2,3:]
```

```python
ax2 = fig.add_subplot(2,1,2)
ax2.plot(years,nuk_f10,label="Fewer than 10 emps")
ax2.plot(years,nuk_m10,label="More than 10 emps")
ax2.plot(years,nuk_any,label="Any number of emps")
ax2.set_title('Number of pubs with different amounts of employees, UK')
ax2.set_xlabel('Year')
ax2.set_ylabel('Number of pubs')
ax2.set_xticks([2000,2005,2010,2015,2020])
ax2.legend()
```
