```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

rng = np.random.RandomState(42)
x = 10 * rng.rand(50)
y = 2 * x - 1 + rng.randn(50)
plt.scatter(x, y);
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression(fit_intercept=True)
X = x[:, np.newaxis] # this reshapes the np array
X.shape
model.fit(X, y)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
# Predicting new values using the regression line 
xfit = np.linspace(-1, 11) # np.linspace returns evenly spaced numbers over a specified interval
Xfit = xfit[:, np.newaxis] # reshaping the axis 
yfit = model.predict(Xfit) # using model.predict is what forecasts 
plt.scatter(x, y)
plt.plot(xfit, yfit) # nice, making a line of regression
```

```python
bigxfit = np.linspace(-1,200) # achieved this by increasing the predictor range using linspace
bigXfit = bigxfit[:, np.newaxis]
yfit = model.predict(bigXfit)
plt.scatter(x, y)
plt.plot(bigXfit, yfit) 
```
